```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
```

```python
# Import image
img = cv2.imread('images/rice.png')

# Apply box filter to image
box_filter_img = cv2.boxFilter(img, -1, (5,5))

# Apply Gaussian filter to image
gaussian_filter = cv2.GaussianBlur(img, (5,5), math.sqrt(2))

# Sharpen images using different filters
sharp_box = np.add(img, np.subtract(img, box_filter_img))
sharp_gauss = np.add(img, np.subtract(img, gaussian_filter))

# Display images
plt.figure(figsize=(20, 20))
plt.subplot(131), plt.imshow(img, cmap = 'gray'), plt.title("Original Image")
plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(sharp_gauss, cmap = 'gray'), plt.title("Sharpened with 5x5 Gaussian Blur")
plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(sharp_box, cmap = 'gray'), plt.title("Sharpened with 5x5 Box Blur")
plt.xticks([]), plt.yticks([])
plt.show()
```

2. Sharpening using box filters as details is expected to give much better results. this is because gaussian filters are a better method for preserving edges, so when you perform the details formula:
*details = original_image - blurred_image*,the details will not be as apparent since the edges have been removed. In this case, the box sharpened looks a bit sharper than the sharpened image that uses Guassian blurring.
